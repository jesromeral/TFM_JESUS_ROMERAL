**Objetivo:**

Construcción de un dataset extendido con variables adicionales de sentimiento.


```python
import pandas as pd
import os
```

```python
SP_PATH   = "../../datos/sp500_model.csv"
SENT_PATH = "../../datos/sentiment_daily_FULL.csv"
OUT_PATH  = "../../datos/sp500_sent_FULL.csv"
```

```python
sp = pd.read_csv(SP_PATH, parse_dates=["Date"])
sent= pd.read_csv(SENT_PATH, parse_dates=["Date"])
```

```python



sp["Date"] = pd.to_datetime(sp["Date"]).dt.normalize()
sent["Date"]   = pd.to_datetime(sent["Date"]).dt.normalize()

sent = sent.groupby("Date", as_index=False).mean(numeric_only=True)


df_sp_sent = sp.merge(sent, on="Date", how="left")

df_sp_sent ["has_news"] = df_sp_sent ["n_news"].notna().astype(int)
df_sp_sent["sentiment_filled"] = df_sp_sent["sentiment_mean"].fillna(0) 
df_sp_sent["n_news_filled"] = df_sp_sent["n_news"].fillna(0).astype(int)

sentiment_cols = [
    "sentiment_median",
    "sentiment_std",
    "pos_ratio",
    "neg_ratio",
    "neu_ratio"
]

df_sp_sent[sentiment_cols] = df_sp_sent[sentiment_cols].fillna(0)

```

```python
print("Filas sp:", len(sp), " | Filas final:", len(df_sp_sent))
print("Rango fechas:", df_sp_sent["Date"].min(), "->", df_sp_sent["Date"].max())
df_sp_sent.head()
```

```python
df_sp_sent.to_csv(OUT_PATH, index=False)

if os.path.exists(OUT_PATH): 
    print(f"Archivo guardado correctamente en {OUT_PATH}") 
else: 
    print("Error: el archivo no se ha guardado.")
```
